```python
from sklearn.cross_validation import train_test_split, cross_val_score
```

```python
stats['MPG'].hist()
```

```python
stats_cols = ['MPG', 'FG', 'FGA', '3P', '3PA', 'FT', 'FTA', 
               'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 
               'TS%', 'eFG%', 'ORB%', 'DRB%', 'TRB%', 
               'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%']
```

```python
X = stats[stats_cols]
```

```python
y = stats['PS/G']
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
Xs = ss.fit_transform(X)
Xs = pd.DataFrame(Xs, columns=X.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xs, y, train_size=0.7)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# use linear regression to predict points per game
# sum up 
lr = LinearRegression().fit(X_train, y_train)
print lr.score(X_train, y_train)
print lr.score(X_test, y_test)
```

```python
ppg = stats.ix[y_test.index,["PS/G"]]
```

```python
linear_ppg = pd.DataFrame(lr.predict(X_test).ravel(), columns=['linear_ppg'], index=y_test.index)
```

```python
ppg = pd.concat([ppg,linear_ppg], axis=1)
```

```python
ppg['Diff'] = ppg['PS/G'] - ppg['linear_ppg']
```

```python
ppg.sort_values('Diff', ascending=False).head()
```

```python
player = pd.DataFrame()
for file in os.listdir("./basketball_reference/game_logs/"):
#     if file.endswith("Curry_Stephen_2017.csv"): # -3.0
    if file.endswith("Iguodala_Andre_2017.csv"): # 0.936
        filepath = "./basketball_reference/game_logs/" + file
#         print filepath
        data = pd.read_csv(filepath)
        player = player.append(data)
```

```python
player_cols = ['mp', 'fg','fga', 'fg3', 'fg3a', 'ft', 'fta',
               'orb', 'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf',
               'ts_pct', 'eFG_pct', 'orb_pct', 'drb_pct', 'trb_pct',
               'ast_pct', 'stl_pct', 'blk_pct', 'tov_pct', 'usg_pct']
# ['pts']
```

```python
player['mp'] = [0 if pd.isnull(item) else round(float(item.split(":")[0]) + float(item.split(":")[1])/60. , 2) for item in player['mp']]
```

```python
ss = StandardScaler()
player_scaled = ss.fit_transform(player[player_cols])
player_scaled = pd.DataFrame(player_scaled, columns=player_cols)
```

```python
lr.score(player_scaled[player_cols], player['pts'])
```

```python
predicted_player = pd.DataFrame(lr.predict(player_scaled[player_cols]), columns=['predicted'])
```
