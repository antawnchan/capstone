```python
def get_vegas_spread(team_name):
    '''
    Given team name from vegas-insiders, extracts spread table from each game
    and saves it to a csv dataframe. Create folder with "./game_results/team_spread/"
    '''
    
    # initiates chrome driver and url
    driver = webdriver.Chrome('/Users/anthony-mba/Downloads/chromedriver')
    url = "http://www.vegasinsider.com/nba/teams/team-page.cfm/team/" + team_name
    driver.get(url)
    
    # locates xpath containing all data
    table_rows = Selector(text=driver.page_source).xpath("//td[@class = 'viBodyBorderNorm']//table/tbody/tr")

    # extracting data from the last table (first two are game scores)
    last_table = Selector(text=driver.page_source).xpath("//table")[-1:]
    rows = last_table.xpath("./tbody/tr")

    # looping through all table rows and saving them into a list
    data_table = []
    for row in rows[1:]: #82 games per season
        try:
            # appending each individual stat to a table_row
            table_row = []
            table_row.append(row.xpath("./td[1]/text()").extract()[0][1:]) # date
            table_row.append(row.xpath("./td[2]/a/text()").extract()[0]) # team name (has hyperlink <a> tag)
            table_row.append(row.xpath("./td[3]/text()").extract()[0].replace(" ","")) # closing line
            table_row.append(row.xpath("./td[4]/text()").extract()[0].replace(" ","")) # total

            # spliting the score data into team and opponent
            score = row.xpath("./td[5]/text()").extract()[0].replace("\t", "").replace("\n", "").replace(u"\xa0", "").replace("Lost", "").replace("Won", "").split("-")
            table_row.append(score[0].replace(" ","")) # team score
            table_row.append(score[1].replace(" ","")) # opponent score
            #     table_row.append(row.xpath("./td[6]/text()").extract()[0].replace("\t", "").replace("\n", "").replace(u"\xa0", "").replace(" ", ""))
            data_table.append(table_row)
        except:
            pass
#         limit to only 82 games a season
#         if len(data_table) == 82:
#             break

    df = pd.DataFrame(data_table, columns=['Date','Opponent','Spread','Over_Under','Team_Score','Opponent_Score'])
    writer = "./game_results/team_spread/" + team_name + ".csv"
    df.to_csv(writer, index=False)
    
    driver.close()
    return data_table
```

# 4) Predict_Player class used to predict points per game (PPG) per player

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
import os
import datetime as dt
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, LassoCV, Ridge, RidgeCV
from sklearn.ensemble import RandomForestRegressor

%matplotlib inline
```

```python
class predict_player:

    # class initialized variables
    df             = None
    opp_df         = None
    cols           = None
    opponents_mask = None
    model          = None
    predicted_ppg  = None
    seed           = None
    
    # training, testing, and seed variables
    X_train        = None
    X_test         = None
    y_train        = None
    y_test         = None
    target_X       = None
    target_y       = None
    Xs             = None
    X              = None
    
    # user initialized variables
    player_url     = None
    opp_team       = None
    last_x_games   = None
    playoff_game   = None
    home_game      = None
    game_started   = None
    player_year    = None
    before_date    = None
    
    def get_parameters(self):
        '''
        Printing all self-initiated or passed in class variables.
        '''
        print "player_url:", self.player_url
        print "opp_team:", self.opp_team
        print "last_x_games:", self.last_x_games
        print "playoff_game:", self.playoff_game
        print "home_game:", self.home_game
        print "game_started:", self.game_started
        print "player_year:", self.player_year
        print "before_date:", self.before_date
    
    def __init__(self, player_url = "Stephen_Curry", opp_team = None, last_x_games = 5, playoff_game = None, home_game = None, game_started = None, player_year = None, before_date = dt.datetime.today()):
        '''
        player_url    : default = "Stephen_Curry"
        opposing_team : default = None, ex: 'SAS'
        last_x_games  : default = 5, last 5 games to set as average prediction. Will always grab last 5 games regardless of playoff_game, home_game, or game_started.
        playoff_game  : default = None, (0,1)
        home_game     : default = None, (0,1)
        game_started  : default = None, (0,1)
        player_year   : default = "", [2017, 2016, 2015, 2014]. Can be list, string, or number.
        before_date   : default = Today, ex: "2017-05-20" = "YYYY-mm-dd"
        '''
        
        #setting up class variables, and performing error/type handling
        self.player_url   = player_url
        if opp_team != None:
            self.opp_team     = opp_team
            self.opp_df       = 0 #################
        if last_x_games != None:
            self.last_x_games = np.abs(last_x_games) * -1
        self.playoff_game = playoff_game
        self.home_game    = home_game
        self.game_started = game_started
        if player_year != None:
#             self.player_year  = str(player_year)
            if type(player_year) == int or type(player_year) == float:
                self.player_year = [str(int(player_year))]
            elif type(player_year) == list:
                self.player_year = [str(x) for x in player_year]
        else: # player_year == None
            self.player_year = [""]
        self.before_date  = before_date
        if type(before_date) == str:
            self.before_date  = dt.datetime.strptime(before_date, "%Y-%m-%d")
        self.df           = self.get_data()
        
        
    
    def predict_player(self):#, opp_team = "GSW", last_x_games = -5, playoff_game = 1, home_game = 1, game_started = 1):
        '''
        Runs through sequential items of class functions to predict player outcome.
        0) Get Data: initialized with class.
        1) Setting train test split.
        2) Setting and defining model.
        3) Setting seed data as average of last_x_games played.
        4) Return results. Returns 0 of not enough games have been played (games played < last_x_games)
        '''
        #initializing global variables
#         if player_url == None: self.player_url = player_url
#         self.opp_team     = opp_team
#         self.last_x_games = np.abs(last_x_games)
#         self.playoff_game = playoff_game
#         self.game_started = game_started
#         self.df           = self.get_data()
        
#         if not self.df.empty:
#         print self.df.shape[0], np.abs(self.last_x_games), self.df.shape[0] > np.abs(self.last_x_games)
        if self.df.shape[0] > np.abs(self.last_x_games):

#             print "setting desired columns for dataframe"
#             cols = list(self.df.columns.values[6:-1])
    #         self.cols = [col for col in cols if col not in ['fg_pct','fg3_pct','ft_pct','pts']] + ['Home_Game', 'playoff']
            self.cols = ['playoff','Home_Game','gs','mp','fga','fg3a','fta','ast','stl','blk','tov','pf','off_rtg']

#             print "setting training data as all data from player"
            self.set_train_test_split()

            self.model = self.get_model(X=self.X_train, y=self.y_train)

    #         return self.get_seed_data(X_test)

#             print "setting seed data based on default X_test"
#             self.seed = self.get_seed_data(X_test=self.X_test, random_data_rows=len(self.y_test))
    #         print self.seed
    
#             print "get player points per game (ppg) prediction"
            self.predicted_ppg = self.get_player_prediction()

            return self.predicted_ppg
        
        else:
            
            return 0.0

    
    def get_data(self):
        '''
        Getting data from all csv files starting with player name and ending with player year.
        Perform data cleaning of csv files.
        '''
        # setting empty dataframe
        df = pd.DataFrame()

        # retreiving data from csv files
        for year in self.player_year:
            for file in os.listdir("./basketball_reference/game_logs/"):
                if file.startswith(self.player_url) and file.endswith(year + ".csv") :
                    filepath = "./basketball_reference/game_logs/" + file
    #                 print filepath
                    data = pd.read_csv(filepath)
                    df = df.append(data)

        df = df.reset_index()

        # deleting unnecessary or columns with NAN
        del df['index']
        del df['fg_pct']
        del df['fg3_pct']
        del df['ft_pct']
        # del df['orb_pct']
        # del df['plus_minus']

        # data cleaning for all game_log files
        df['date'] = pd.to_datetime(df['date'])
        df.rename(columns={'loc': 'Home_Game'}, inplace=True)
        df['playoff'] = [0 if item[-1] == "N" else 1 for item in df['playoff']]
        df['Home_Game'] = [0 if item == "@" else 1 for item in df['Home_Game']]
        df['result'] = [item[3:].replace(")","") for item in df['result']]
        df['result'] = df['result'].astype(float)
        df['mp'] = [0 if pd.isnull(item) else round(float(item.split(":")[0]) + float(item.split(":")[1])/60. , 2) for item in df['mp']]
        
        df = df.dropna(axis=0, how='any')
        df = df.sort_values('date')
        df = df[df['date'] < self.before_date]
        return df
    
    
#     def get_defensive_stats():
#         defense = pd.read_csv("./basketball_reference/team_defensive_rating.csv")
#         defense[defense["short_name"] == self.opp_team][]
    
    def set_train_test_split(self):#, opp_team, last_x_games, playoff_game, df, cols):
        '''
        Setting opponent information as average seed data of last_x_games based on inputs of opponent, home game, playoff game, and date.
        Used Standard Scalar to scale data.
        Setting train test split as standand options, or first and last games played up to a particular date.
        '''
        
        # setting opponent information based on past x games
        opp = self.df[self.last_x_games:] 
        if self.opp_team != None:
            opp = self.df[(self.df['opp'] == self.opp_team)]
            try:
                opp = opp[self.last_x_games:]
            except e: pass # catch index out of range: if last_x_games > game played per opponent
        
        # setting opponent information based on playoff data
        playoff = pd.DataFrame()
        if self.playoff_game != None:
            try:
                if self.playoff_game == 1 or self.playoff_game == 0:
                    playoff = self.df[(self.df['playoff'] == self.playoff_game)][self.last_x_games:]
            except e: pass
        
        # setting opponent information based on home game data
        home = pd.DataFrame()
        if self.home_game != None:
            home = self.df[(self.df['Home_Game'] == self.home_game)][self.last_x_games:]
        
        
        opponents = pd.concat([opp, playoff, home])#['pts'] # df['playoff'] == playoff_game
        # setting testing data based on team opponent and last 5 games
        mask = set(opponents.index.values) # opponents.drop_duplicates()
        self.opponents_mask = mask
        
        not_mask = [i for i in self.df.index if i not in mask]
        
        # setting training data as all data from player
#         X_train = self.df.ix[not_mask,:][self.cols]
#         y_train = self.df.ix[not_mask,:][['pts']]
#         X_test = self.df.ix[mask,:][self.cols]
#         y_test = self.df.ix[mask,:][['pts']]
        
#         X = self.df.ix[not_mask,:][self.cols]
#         y = self.df.ix[not_mask,:][['pts']]
        
        X = self.df[self.cols]
        self.X = X
        y = self.df[['pts']]
        
        Xs = StandardScaler().fit_transform(X)
        Xs = pd.DataFrame(Xs, columns=self.cols, index=X.index)
        self.Xs = Xs
        
        # tts
        train_size = 0.9
#         X_train, X_test, y_train, y_test = train_test_split(Xs, y, train_size=train_size, random_state=8)
#         print "mp: ", X_train['mp'].mean(), X_test['mp'].mean()
#         print "pts: ", y_train['pts'].mean(), y_test['pts'].mean()

        # early and late game split
        X_train = Xs[self.cols].head(int(round(Xs.shape[0] * train_size)))
        y_train = y[['pts']].head(int(round(Xs.shape[0] * train_size)))
        X_test = Xs[self.cols].tail(int(round(Xs.shape[0] * (1. - train_size))))
        y_test = y[['pts']].tail(int(round(Xs.shape[0] * (1. - train_size))))

        self.X_train = X_train
        self.X_test  = X_test
        self.y_train = y_train
        self.y_test  = y_test
        self.target_X = Xs.ix[mask,:][self.cols]#.dropna()
        self.target_y = y.ix[mask,:][['pts']]#.dropna()
        
#         return (X_train, X_test, y_train, y_test)
    
    
    def get_model(self, X, y):
        '''
        Set up model with training data.
        '''
        model = RidgeCV().fit(X, y)
#         model = RandomForestRegressor(max_depth=3).fit(X, y)
#         print "score X_train:", model.score(self.X_train, self.y_train)
#         print "score X_test:", model.score(self.X_test, self.y_test)
#         print "score target_X:", model.score(self.target_X, self.target_y)
        return model
    
    def score(self, X, y):
        return self.model.score(X, y)
        
    
    def get_seed_data(self, X_test, random_data_rows=1):
        '''
        Not used: generating seed data to as average of last_x_games or by random Gaussian.
        '''
        # randomly seeding data based on mean/std from a given opponent
        seed = []
        for i in range(random_data_rows):
            row = []
            for col in self.cols:
                if col == "Home_Game":
                    if self.home_game != None:
                        row.append(self.home_game) # home game == 1, away game == 0
                    else: 
                        row.append(int(round(X_test['Home_Game'].mean())))
#                         row.append(np.random.choice([0,1]))
                elif col == 'gs':
                    if self.game_started != None:
                        row.append(self.game_started)
                    else:
                        row.append(int(round(X_test['gs'].mean())))
#                         row.append(X_test['gs'].value_counts().idxmax()) # max of started vs not started
                elif col == 'playoff':
                    if self.playoff_game != None:
                        row.append(self.playoff_game)
                    else:
                        row.append(int(round(X_test['playoff'].mean())))
#                         row.append(X_test['playoff'].value_counts().idxmax())
                else:
                    std = X_test[col].std()
                    if std == 0:
                        row.append(0)
                    else:
                        row.append(X_test[col].mean())
    #                     row.append(np.random.normal(X_test[col].mean(), X_test[col].std()))
            seed.append(row)

    #     print ridge.score(X_test, y_test)
        seed = pd.DataFrame(seed, columns=self.cols)
        return seed
    
    def predict(self, X):
        return self.model.predict(X)
    
    def get_player_prediction(self):
        '''
        target_X is the average seed data of last_x_games based on inputs of opponent, home game, playoff game, and date.
        '''
        return self.predict(X=self.target_X).mean()
    
```

```python
# returns a weighted average team score for one game of 240 minutes based on a given list of players
def get_ppg_from_list(players_list, opp_team, last_x_games, playoff_game, home_game,\
                      player_year, before_date=dt.datetime.today()):
    game_score = []
    for player in players_list:
        # call predict_player class with input parameters
        pp = predict_player(player_url = player, opp_team=opp_team, last_x_games=last_x_games,\
                            playoff_game=playoff_game, home_game=home_game, player_year=player_year,\
                            before_date=dt.datetime.today())
        pp.predict_player()
        
        # set up array: player name, average points in last_x_games, predicted points, and average minutes played
        game_score.append([player, pp.target_y['pts'].mean(), pp.model.predict(pp.target_X).mean(),\
                           pp.df.ix[pp.opponents_mask,:]['mp'].mean()])
    # import array to dataframe
    game_score = pd.DataFrame(game_score, columns=['player','orig','ppg','mpg'])
    
    # return weighted average of team score
    return game_score.ppg.sum() * (48*5.) / game_score.mpg.sum()
```
